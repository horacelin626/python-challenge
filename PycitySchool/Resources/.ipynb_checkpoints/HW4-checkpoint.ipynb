```python
import pandas as pd
import numpy as np
```

```python
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"
```

```python
school_data_pd = pd.read_csv(school_data_to_load)
student_data_pd = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data_pd, school_data_pd, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
total_number_school= school_data_pd["school_name"].count()

total_number_students = student_data_pd["student_name"].count()

total_budget = school_data_pd["budget"].sum()

average_math_score = student_data_pd["math_score"].mean()

average_reading_score = student_data_pd["reading_score"].mean()

overall_passing_rate = (average_math_score + average_reading_score)/2

passing_math = ((school_data_complete["math_score"])>=70).mean() *100

passing_read =((school_data_complete["reading_score"])>=70).mean()*100
```

```python
summary_table = pd.DataFrame({"Total Schools": [total_number_school], "Total Student":[total_number_students],
                              "Total Burdget":[total_budget], "Average Math Score":[average_math_score],
                             "Average Reading Score":[average_reading_score] ,"% Passing Math":[passing_math],
                             "% Passing Reading":[passing_read],"% Overall Passing Rate":[overall_passing_rate]})

summary_table
```

```python
#School Summary
```

```python
school_type = school_data_pd.set_index(["school_name"])["type"]

total_student = school_data_complete["school_name"].value_counts()

total_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]

per_student_budget = total_budget / total_student

ave_math_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]

ave_read_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]

passing_read = school_data_complete[(school_data_complete["reading_score"] >= 70)]

passing_math_rate = passing_math.groupby(["school_name"]).count()["Student ID"] / total_student

passing_read_rate = passing_read.groupby(["school_name"]).count()["Student ID"] / total_student

overall_passing_rate = (passing_math_rate + passing_read_rate ) / 2


School_summary_table = pd.DataFrame({"School Type":school_type, "Total Student":total_student,
                                   "Ttoal School Budget":total_budget, "Per Student Budget":per_student_budget,
                                   "Average Math Score":ave_math_score,"Average Reading Score":ave_read_score,
                                   "% Passing Math":passing_math_rate,"% Passing Reading":passing_read_rate,
                                   "% Overall Passing Rate":overall_passing_rate})


School_summary_table
```

```python
top_schools = School_summary_table.sort_values(["% Overall Passing Rate"], ascending=False)
top_schools.head(5)
```

```python
bottom_schools = School_summary_table.sort_values(["% Overall Passing Rate"], ascending=True)
bottom_schools.head(5)
```

```python
# Math Scores by Grade
ninth_grade = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_grade = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_grade = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_grade = school_data_complete[(school_data_complete["grade"] == "12th")]

ninth_grade_score = ninth_grade.groupby(["school_name"]).mean()["math_score"]
tenth_grade_score = tenth_grade.groupby(["school_name"]).mean()["math_score"]
eleventh_grade_score = eleventh_grade.groupby(["school_name"]).mean()["math_score"]
twelfth_grade_score = twelfth_grade.groupby(["school_name"]).mean()["math_score"]

scores_by_grade = pd.DataFrame({"9th": ninth_grade_score, "10th": tenth_grade_score,
                                "11th": eleventh_grade_score, "12th": twelfth_grade_score})


scores_by_grade
```

```python
# Reading Scores by Grade
ninth_grade = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_grade = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_grade = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_grade = school_data_complete[(school_data_complete["grade"] == "12th")]

ninth_grade_score = ninth_grade.groupby(["school_name"]).mean()["reading_score"]
tenth_grade_score = tenth_grade.groupby(["school_name"]).mean()["reading_score"]
eleventh_grade_score = eleventh_grade.groupby(["school_name"]).mean()["reading_score"]
twelfth_grade_score = twelfth_grade.groupby(["school_name"]).mean()["reading_score"]

scores_by_grade = pd.DataFrame({"9th": ninth_grade_score, "10th": tenth_grade_score,
                                "11th": eleventh_grade_score, "12th": twelfth_grade_score})


scores_by_grade

```

```python
#Scores by School Spending
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

School_summary_table["Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)

spending_math_scores = School_summary_table.groupby(["Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = School_summary_table.groupby(["Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = School_summary_table.groupby(["Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = School_summary_table.groupby(["Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_rate = (spending_passing_math + spending_passing_reading) / 2

spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})

spending_summary
```

```python
#Scores by School Size

size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

School_summary_table["size"] = pd.cut(total_student, size_bins, labels=group_names)

size_math_scores = School_summary_table.groupby(["size"]).mean()["Average Math Score"]
size_reading_scores = School_summary_table.groupby(["size"]).mean()["Average Reading Score"]
size_passing_math = School_summary_table.groupby(["size"]).mean()["% Passing Math"]
size_passing_reading = School_summary_table.groupby(["size"]).mean()["% Passing Reading"]
overall_passing_rate = (size_passing_math + size_passing_reading) / 2

size_summary = pd.DataFrame({"Average Math Score" : size_math_scores,
                                 "Average Reading Score": size_reading_scores,
                                 "% Passing Math": size_passing_math,
                                 "% Passing Reading": size_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})

size_summary
```

```python
#Scores by School Type
type_math_scores = School_summary_table.groupby(["School Type"]).mean()["Average Math Score"]
type_reading_scores = School_summary_table.groupby(["School Type"]).mean()["Average Reading Score"]
type_passing_math = School_summary_table.groupby(["School Type"]).mean()["% Passing Math"]
type_passing_reading = School_summary_table.groupby(["School Type"]).mean()["% Passing Reading"]
overall_passing_rate = (type_passing_math + type_passing_reading) / 2

# Assemble into data frame
type_summary = pd.DataFrame({"Average Math Score" : type_math_scores,
                             "Average Reading Score": type_reading_scores,
                             "% Passing Math": type_passing_math,
                             "% Passing Reading": type_passing_reading,
                             "% Overall Passing Rate": overall_passing_rate})

type_summary

```
